# PacketFormer — Phase 1 Baseline Exploration

This notebook explores the CICIDS2017 dataset before training.
Run `python src/preprocess.py` first to generate `data/processed/`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

sns.set_theme(style='darkgrid')
```

```python
X = np.load('data/processed/X.npy')
y = np.load('data/processed/y.npy')

with open('data/processed/label_encoder.pkl', 'rb') as f:
    le = pickle.load(f)

print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
print(f'Classes: {le.classes_}')
```

```python
# Class distribution
labels, counts = np.unique(y, return_counts=True)
label_names = le.inverse_transform(labels)

plt.figure(figsize=(10, 5))
sns.barplot(x=label_names, y=counts)
plt.yscale('log')
plt.xticks(rotation=30, ha='right')
plt.title('CICIDS2017 Class Distribution (log scale)')
plt.tight_layout()
plt.show()
```
